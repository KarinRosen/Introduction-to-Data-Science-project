```python
import pandas as pd
```

```python
data2017 = pd.read_csv('RV36U_2017keskellega.csv')
data2018 = pd.read_csv('RV36U_2018keskellega.csv')
data2019 = pd.read_csv('RV36U_2019keskellega.csv')
data2020 = pd.read_csv('RV36U_2020keskellega.csv')
data2021 = pd.read_csv('RV36U_2021keskellega.csv')
```

```python
data2017 = data2017.drop(columns = 'Aasta')
data2018 = data2018.drop(columns = 'Aasta')
data2019 = data2019.drop(columns = 'Aasta')
data2020 = data2020.drop(columns = 'Aasta')
data2021 = data2021.drop(columns = 'Aasta')
```

```python
data2017
```

```python
data = pd.concat([data2017, data2018, data2019, data2020, data2021]).groupby(['Meeste vanuser端hm']).sum().reset_index()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = data.set_index('Meeste vanuser端hm')
```

```python
fig, ax = plt.subplots(figsize=(15,13))
heatmap = sns.heatmap(data, annot=True,linewidth=.5, fmt="d", cmap = 'RdPu')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
heatmap.xaxis.tick_top()
```

```python

results_path = 'lahutused_heatmap'
heatmap.figure.savefig(results_path, dpi=500, bbox_inches='tight')
```
