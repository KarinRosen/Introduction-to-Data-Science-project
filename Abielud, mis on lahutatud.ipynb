```python
import pandas as pd
```

```python
abielud = pd.read_csv("abielud_kuud_maakonnad.csv")
abielud.query("Aasta == 2017")
```

```python
lahutused = pd.read_csv("abielu_pikkus.csv")

abielu_kestus_M = lahutused.query("Aasta == 2017 and Sugu == 'Mehed'")[['Alla 1 aasta',
       '1 aasta', '2 aastat', '3 aastat', '4 aastat', '5 aastat', '6 aastat',
       '7 aastat', '8 aastat', '9 aastat', '10-14 aastat', '15-19 aastat',
       '20 ja enam aastat', 'Abielu kestus teadmata']].sum()
meeste_arv = abielu_kestus_M.sum()

abielu_kestus_N = lahutused.query("Aasta == 2017 and Sugu == 'Naised'")[['Alla 1 aasta',
       '1 aasta', '2 aastat', '3 aastat', '4 aastat', '5 aastat', '6 aastat',
       '7 aastat', '8 aastat', '9 aastat', '10-14 aastat', '15-19 aastat',
       '20 ja enam aastat', 'Abielu kestus teadmata']].sum()
naiste_arv = abielu_kestus_N.sum()

abielu_kestus_M = abielu_kestus_M.to_list()
abielu_kestus_N = abielu_kestus_N.to_list()
```

```python
mehed_naised = pd.DataFrame({'Kestus': ['Alla 1 aasta',
       '1 aasta', '2 aastat', '3 aastat', '4 aastat', '5 aastat', '6 aastat',
       '7 aastat', '8 aastat', '9 aastat', '10-14 aastat', '15-19 aastat',
       '20 ja enam aastat', 'Abielu kestus teadmata'],
                             'Mehed': abielu_kestus_M,
                             'Naised': abielu_kestus_N 
                             })

mehed_naised['Mehed'] /= meeste_arv
mehed_naised['Naised'] /= naiste_arv
```

```python
mehed_naised.plot(grid=True, x="Kestus", y=["Mehed", "Naised"], kind="bar",)
```
