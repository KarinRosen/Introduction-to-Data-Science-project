```python
import pandas as pd 
import numpy as np
pikkused = pd.read_csv("õigetabel_abielu_pikkus_lahutuseni.csv")
```

```python
pikkused
totalid = pikkused['Total']
asendus = []
for item in totalid:
    item = int(item.replace(',', ''))
    asendus.append(item)
pikkused
pikkused["Total"] = asendus
pikkused["Total"] = pikkused["Total"]-pikkused["Unknown"]
```

```python
pikkused = pikkused.drop(columns= ["Unknown"])
pikkused = pikkused.rename(columns= {"DURATION (Labels)" : "Aasta"})
pikkused.set_index('Aasta', inplace=True)
```

```python
pikkused = pikkused.rename(columns= {"Total" : "Kokku"})
pikkused = pikkused.drop(columns= ["Kokku"])
pikkused = pikkused.rename(columns= {"Less than 1 year" : 0})
pikkused = pikkused.rename(columns= {"1 year" : 1})
pikkused = pikkused.rename(columns= {"35 years or over":35})
```

```python
for i in range(2, 35):
    pikkused = pikkused.rename(columns= {str(i)+" years":int(i)})
pikkused
```

```python
pikkused
```

```python

m = list()
for index, row in pikkused.iterrows():
    v = list() # value list
    z = zip(row.index, row.values)
    for item in z:
        for f in range(item[1]):
            v.append(item[0])
    m.append(np.median(v))
df_m = pd.DataFrame({'Aasta': pikkused.index, 'Mediaan': m})
print(df_m)
```

```python
df_m.plot.line(x="Aasta", y= "Mediaan", xlabel= "Lahutuste abielude pikkuste mediaani muutumine ajas", figsize =(20, 10) )
```
