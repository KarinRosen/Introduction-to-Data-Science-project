```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sots_lahutus = pd.read_csv("sots_lahutus.csv")
sots_abielu = pd.read_csv("sots_abielud.csv")
```

```python
sots_abielu = sots_abielu.groupby(["Aasta", "Sugu"]).sum()
sots_lahutus = sots_lahutus.groupby(["Aasta", "Sugu"]).sum()
```

```python
abielu_veerud = ['Töötav','Töötu','Pensionär','(Üli)õpilane','Kodune']
sots_abielu['Kokku'] = sots_abielu[abielu_veerud].sum(axis=1)
lahutus_veerud = ['Töötav','Töötu','Pensionär','(Üli)õpilane','Kodune']
sots_lahutus['Kokku'] = sots_lahutus[lahutus_veerud].sum(axis=1)

```

```python
abielu_kokku_mehed_naised = sots_abielu.droplevel(1)
abielu_kokku_mehed_naised =abielu_kokku_mehed_naised.groupby("Aasta").sum()
lahutus_kokku_mehed_naised = sots_lahutus.droplevel(1)
lahutus_kokku_mehed_naised = lahutus_kokku_mehed_naised.groupby("Aasta").sum()
```

```python
sots_abielu_mehed = sots_abielu.drop('Naised',level = 1)
sots_abielu_mehed['Töötu %'] =sots_abielu_mehed['Töötu']/abielu_kokku_mehed_naised['Kokku']
sots_abielu_naised = sots_abielu.drop('Mehed',level = 1)
sots_abielu_naised['Töötu %'] =sots_abielu_naised['Töötu']/abielu_kokku_mehed_naised['Kokku']
```

```python
sots_lahutus_mehed = sots_lahutus.drop('Naised',level = 1)
sots_lahutus_mehed['Töötu %'] =sots_lahutus_mehed['Töötu']/lahutus_kokku_mehed_naised['Kokku']
sots_lahutus_naised = sots_lahutus.drop('Mehed',level = 1)
sots_lahutus_naised['Töötu %'] =sots_lahutus_naised['Töötu']/lahutus_kokku_mehed_naised['Kokku']
```

```python
sots_abielu_mehed = sots_abielu_mehed.droplevel(1) 
sots_abielu_naised = sots_abielu_naised.droplevel(1)
sots_lahutus_mehed = sots_lahutus_mehed.droplevel(1) 
sots_lahutus_naised = sots_lahutus_naised.droplevel(1) 
```

```python
f = plt.figure()
f.set_figwidth(30)
f.set_figheight(10)
plt.plot(sots_abielu_mehed["Töötu %"],color = "purple", marker='o', markerfacecolor='black', label="Mehed", linewidth=4)
plt.plot(sots_abielu_naised["Töötu %"],color = "pink", marker='o', markerfacecolor='black', label="Naised", linewidth=4)
plt.legend()
plt.show()
```

```python
f = plt.figure()
f.set_figwidth(30)
f.set_figheight(10)
plt.plot(sots_lahutus_mehed["Töötu %"],color = "purple", marker='o', markerfacecolor='black', label="Mehed", linewidth=4)
plt.plot(sots_lahutus_naised["Töötu %"],color = "pink", marker='o', markerfacecolor='black', label="Naised", linewidth=4)
plt.legend()
plt.show()
```
