```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors
```

```python
abielud = pd.read_csv('RV27_mitmesabielu.csv')
lahutused = pd.read_csv('RV29_mitmeslahutus.csv')
```

```python
abielud
```

```python
abielud["Esmakordselt abiellunud"] = abielud['Esmakordselt abiellunud Mehed'].astype(int) + abielud["Esmakordselt abiellunud Naised"]
```

```python
abielud["Korduvalt abiellunud"] = abielud['Korduvalt abiellunud Mehed'].astype(int) + abielud["Korduvalt abiellunud Naised"]
```

```python
abielud["Abielud kokku"] = abielud['Korduvalt abiellunud'].astype(int) + abielud["Esmakordselt abiellunud"]
```

```python
abielud["Abielud kokku, Mehed"] = abielud['Korduvalt abiellunud Mehed'].astype(int) + abielud["Esmakordselt abiellunud Mehed"]
```

```python
abielud["Abielud kokku, Naised"] = abielud['Korduvalt abiellunud Naised'].astype(int) + abielud["Esmakordselt abiellunud Naised"]
```

```python
abielud['Korduvabielude osatähtsus %'] = abielud['Korduvalt abiellunud'] / abielud['Abielud kokku'] * 100
```

```python
abielud['Korduvabielude osatähtsus %, Mehed'] = abielud['Korduvalt abiellunud Mehed'] / abielud['Abielud kokku, Mehed'] * 100
```

```python
abielud['Korduvabielude osatähtsus %, Naised'] = abielud['Korduvalt abiellunud Naised'] / abielud['Abielud kokku, Naised'] * 100
```

```python
abielud[['Korduvabielude osatähtsus %, Mehed', 'Korduvabielude osatähtsus %, Naised']].mean()
```

```python
abielud.plot.line(y = ['Korduvabielude osatähtsus %', 'Korduvabielude osatähtsus %, Mehed', 'Korduvabielude osatähtsus %, Naised'], x = 'Aasta', figsize = (10,10))
```

```python
lahutused["Lahutanud Mehed"] = lahutused['Korduvabielu lahutanud Mehed'].astype(int) + lahutused["Esmasabielu lahutanud Mehed"]
```

```python
lahutused["Lahutanud Naised"] = lahutused['Korduvabielu lahutanud Naised'].astype(int) + lahutused["Esmasabielu lahutanud Naised"]
```

```python
lahutused['Korduvabielude lahutuste osatähtsus %'] = lahutused['Korduvabielu lahutanud'] / lahutused['Lahutused kokku'] * 100
```

```python
lahutused['Korduvabielude lahutuste osatähtsus %, Mehed'] = lahutused['Korduvabielu lahutanud Mehed'] / lahutused['Lahutanud Mehed'] * 100
```

```python
lahutused['Korduvabielude lahutuste osatähtsus %, Naised'] = lahutused['Korduvabielu lahutanud Naised'] / lahutused['Lahutanud Naised'] * 100
```

```python
lahutused
```

```python
lahutused[['Korduvabielude lahutuste osatähtsus %, Mehed', 'Korduvabielude lahutuste osatähtsus %, Naised']].mean()
```

```python
lahutused.plot.line(y = ['Korduvabielude lahutuste osatähtsus %, Mehed', 'Korduvabielude lahutuste osatähtsus %, Naised'], x = 'Aasta', figsize = (10,10))
```

```python
ax = lahutused.plot(kind = 'line', x = 'Aasta',
                  y = ['Korduvabielude lahutuste osatähtsus %, Mehed', 'Korduvabielude lahutuste osatähtsus %, Naised'], color = 'Blue',
                  linewidth = 3, figsize = (10,10))
ax2 = lahutused.plot(kind = 'line', x = 'Aasta',
                   y = 'Lahutused kokku', secondary_y = True,
                   color = 'Red',  linewidth = 3,
                   ax = ax)
plt.show()
#ax = lahutused.plot(x = 'Aasta', )
#plt.title('adjkd')

#ax2 = ax.twinx()
#ax.plot(y1 = ['Korduvabielude lahutuste osatähtsus %', 'Korduvabielude lahutuste osatähtsus %, Mehed', 'Korduvabielude lahutuste osatähtsus %, Naised'])
#ax2.plot(y2 = ['Lahutused kokku'])
```
