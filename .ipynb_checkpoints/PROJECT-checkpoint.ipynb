```python
import pandas as pd
```

```python
sots = pd.read_csv("sots.csv")
pikkus = pd.read_csv("abielu_pikkus.csv")
maakond = pd.read_csv("maakond.csv")
kuud = pd.read_csv("kuud_maakonnad.csv")
```

```python
sots['Elukoht'] = sots['Elukoht'].str.replace('..linnaline ja väikelinnaline asustuspiirkond\xa0\xa0','Linnaline ja väikelinnaline asustuspiirkond')
sots['Elukoht'] = sots['Elukoht'].str.replace('..maaline asustuspiirkond','Maaline asustuspiirkond')

pikkus['Elukoht'] = pikkus['Elukoht'].str.replace('linnaline ja väikelinnaline asustuspiirkond','Linnaline ja väikelinnaline asustuspiirkond')
pikkus['Elukoht'] = pikkus['Elukoht'].str.replace('maaline asustuspiirkond','Maaline asustuspiirkond')

sots = sots.rename(columns={"Elukoht":"Asustuspiirkond"})
pikkus = pikkus.rename(columns={"Elukoht":"Asustuspiirkond"})
maakond = maakond.rename(columns= {"Asustuspiirkonna liik" : "Asustuspiirkond"})
```

```python
pikkus2017 = pikkus.query("Aasta == 2017 & Sugu == 'Mehed'")
pikkus2017 = pikkus2017.drop(columns=["Sugu", "Aasta"])

sots2017 = sots.query("Sugu == 'Mehed'")
sots2017 = sots2017.drop(columns=["Sugu", "Aasta"])
```

```python
kokku = sots2017.merge(pikkus2017, on=["Asustuspiirkond", "Vanuserühm", "Kokku"])

maaline = kokku.query("Asustuspiirkond == 'Maaline asustuspiirkond'")
linnaline = kokku.query("Asustuspiirkond == 'Linnaline ja väikelinnaline asustuspiirkond'")
```

```python
maakond['Mehe elukoht'] = maakond['Mehe elukoht'].str.replace('Harju maakond','Harju maakond (v.a Tallinn)')
kuud['Mehe elukoht'] = kuud['Mehe elukoht'].str.replace('Harju maakond','Harju maakond (v.a Tallinn)')
maakond = maakond.rename(columns={"Mehe elukoht":"Elukoht"})
kuud = kuud.rename(columns={"Mehe elukoht":"Elukoht"})
```
